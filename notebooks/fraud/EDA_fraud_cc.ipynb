```python
data_path = r"C:\GitHub_Master\ML---Financial-Risk-Intelligence-Platform\data\raw\Fraud Detection\Credit Card\credit_card_transactions.csv"
df = pd.read_csv(data_path)
df.head()
```

```python
columns_to_drop = [
    "Unnamed: 0",
    "first",
    "last",
    "street",
    "city",
    "state",
    "zip",
    "trans_num",
    "unix_time",
    "merch_zipcode"
]

df.drop(columns=columns_to_drop, inplace=True, errors="ignore")
df.head()
```

```python
df['is_fraud'].value_counts(normalize=True)
```

```python
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['dob'] = pd.to_datetime(df['dob'])
```

```python
# Time features
df['hour'] = df['trans_date_trans_time'].dt.hour
df['day'] = df['trans_date_trans_time'].dt.day
df['month'] = df['trans_date_trans_time'].dt.month
df['year'] = df['trans_date_trans_time'].dt.year
df['dayofweek'] = df['trans_date_trans_time'].dt.dayofweek

# Age (accurate)
df['card_holder_age'] = (df['trans_date_trans_time'] - df['dob']).dt.days // 365

# Drop dob
df.drop(columns=["dob"], inplace=True)
```

1. Target Variable – Fraud Distribution

```python
SAVE_DIR = r"C:\GitHub_Master\ML---Financial-Risk-Intelligence-Platform\reports\fraud\creditcard"
os.makedirs(SAVE_DIR, exist_ok=True)

# Helper function to show and save plot
def show_and_save(fig, filename):
    filepath = os.path.join(SAVE_DIR, filename)
    fig.savefig(filepath, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = sns.countplot(x='is_fraud', data=df, palette='viridis')

# Add annotations
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height()):,}',                     # formatted count
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='bottom',
        fontsize=11, color='black'
    )

plt.title('Distribution of Fraud vs Non-Fraud Transactions', fontsize=14)
plt.xlabel('is_fraud', fontsize=12)
plt.ylabel('Count', fontsize=12)

show_and_save(fig, "is_fraud_distribution.png")
```

```python
categorical_columns = ['category', 'gender']

for column in categorical_columns:
    plt.figure(figsize=(12, 6))
    ax = sns.countplot(y=column, hue='is_fraud', data=df, palette='viridis')
    plt.title(f'{column} Distribution by is_fraud')
    plt.xlabel('Count')
    plt.ylabel(column)
    plt.legend(title='credit approval', loc='upper right', bbox_to_anchor=(1.15, 1))
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(True)
    ax.spines['bottom'].set_visible(True)
    
    for p in ax.patches:
        width = p.get_width()
        ax.text(width + 10, p.get_y() + p.get_height() / 2, 
                f'{int(width)}', 
                ha='left', va='center')

    show_and_save(fig, "Distribution by is_fraud.png")
```

```python
num_cols = df.select_dtypes(include=np.number).columns

df[num_cols].hist(figsize=(16, 16), bins=30)
plt.tight_layout()
plt.show()
plt.savefig(os.path.join(REPORT_PATH, "numeric_histograms.png"))
```

```python
cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(y=df[col])
    plt.title(f"Distribution of {col}")
    plt.savefig(os.path.join(REPORT_PATH, f"{col}_distribution.png"))
    plt.show()
```
